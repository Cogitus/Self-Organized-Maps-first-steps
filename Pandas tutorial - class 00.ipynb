Posso descobrir o número de valores únicos de uma dada coluna (Series) usando o método .value_counts() - é uma excelente forma de entender a distribuição de seus dados

```python
degree_counts = df['Level of Education'].value_counts()
degree_counts_index = df['Level of Education'].value_counts().keys().tolist()

degree_counts
```

```python
import itertools #preciso para poder iterar sobre duas listas simultaneamentes!

for quantidade, indices in zip(degree_counts, degree_counts_index):
    print('Quantidade de elementos de {}: {};'.format(indices, quantidade))
```

Uma peculiaridade da biblioteca pandas é que também consigo plotar diretamente do objeto em questão sem ter que chamar outras bibliotecas parar plotar pra mim! Veja:

```python
degree_counts.plot(kind='bar') #gráfico de frequência de repetições (plotado por um método do objeto Series)
```
