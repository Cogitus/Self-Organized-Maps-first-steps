```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#seeing how a uniform distribution works.
someValues = np.random.uniform(-100, 100, 50000) #there are 50000 thousand values in [-10;10]

plt.hist(someValues, bins=1000, density=True, histtype='stepfilled') #bins=1000 -> 1000 edges!
plt.show()
```

```python
from scipy.stats import norm

x = np.arange(-5, 5, 0.001)
plt.plot(x, norm.pdf(x)) #norm.pdf() -> probability distribution function for a normal distribution!

#note that we used the following syntax: plt.plot(x, f(x))! the only "different" thing here is how we calculate f(x)=y
```

```python
#an exponentital distribution function
from scipy.stats import expon

#reusing 'x' for a new distribution
x = np.arange(-10, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

```python
#seeing the behavior of a poisson distribution
from scipy.stats import poisson

mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
